```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import skewnorm
```

```python
a1=4.0
```

```python
a2=10.0
```

```python
dist1=skewnorm(a1)
```

```python
dist2=skewnorm(a2)
```

```python
x=np.linspace(skewnorm.ppf(0.01, a1), skewnorm.ppf(0.99, a1), 100)
```

```python
plt.plot(x,dist1.pdf(x))
```

```python
plt.plot(x,dist2.pdf(x))
```
